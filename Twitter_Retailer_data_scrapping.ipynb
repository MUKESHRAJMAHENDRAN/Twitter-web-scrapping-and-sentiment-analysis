### Extracting the tweets with respect the ebay, Meesho, Olxindia, Amazonindia, Flipkart, Jiomart

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Creating list to append tweet data to
ebay_list = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('ebay since:2019-01-01 until:2021-12-27').get_items()):
    if i>10000:
        break
    ebay_list.append([tweet.date, tweet.id, tweet.username, tweet.content])
    
# Creating a dataframe from the tweets list above
ebay_df = pd.DataFrame(ebay_list, columns=['Datetime', 'Tweet Id', 'Name', 'tweet'])
```
